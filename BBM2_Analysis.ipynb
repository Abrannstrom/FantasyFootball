```python
# Add Matplotlib inline magic command
#%matplotlib inline
#get_ipython().run_line_magic('matplotlib', 'inline')
# Dependencies and Setup
#import matplotlib.pyplot as plt
import pandas as pd
from pandas import * 
import numpy as np
```

```python
# 1. Create a DataFrame for the bestball reg season data
BBM2_data = "Resources/BBM2_2021.csv"
BBM2_df = pd.read_csv(BBM2_data)
```

```python
# 2. Check our columns
BBM2_df.info()
```

```python
BBM2_df.head()
```

```python
BBM2_df = BBM2_df.drop(['draft_id', 'draft_time', 'clock', 'tournament_entry_id', 'bye_week'], axis=1)
```

```python
BBM2_df = BBM2_df.rename(columns={"team_pick_number": "round", "pick_points": "points_added", "projection_adp": "adp"})
```

```python
#Find advance rates
df = BBM2_df.groupby(["player_name"]).mean()["playoff_team"]
df
```

```python
df2 = BBM2_df.groupby(["player_name"]).mean()["adp"]
df2
```

```python
df3 = BBM2_df.groupby(["player_name"])
df3
```

```python
df4 = BBM2_df.groupby(["player_name"]).mean()["points_added"]
df4
```

```python
df6 = BBM2_df.groupby(["player_name"]).mean()["round"]
df6
```

```python
df7 = BBM2_df.groupby(["player_name", "position_name"])
df7
```

```python
#  6. Create a PyBer summary DataFrame. 
summary_df = pd.DataFrame({
          #"Player Name" : df3,
          #"Position": BBM2_df(["position_name"]),
          "ADP": df2,
          "Round": df6,
          "Advance Rate": df,
          "Points Added": df4})

summary_df
```

```python
df5 = BBM2_df.groupby(["player_name", "position_name", "round"]).mean()["playoff_team"]
df5
```
